```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
#sentences
sent= [
    'I love my dog',
    'I love my cat',
    'You love my dog!',
    'Do you think my dog is amazing?'
]
```

```python
sent
```

```python
# define the vocuabulary size
voc_size=10000
```

```python
#  one hot representation
one_hot_representations = [one_hot(words, voc_size) for words in sent]
one_hot_representations
```

```python
# word embedding representation 
from tensorflow.keras.layers import Embedding , Input
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
# pad the sequences to have same length
sent_length=8
embedded_docs=pad_sequences(one_hot_representations,maxlen=sent_length,padding='pre')
print(embedded_docs)
```

```python
#  feature representation 
dim = 10
```

```python
model = Sequential()
model.add(Input(shape=(sent_length,)))
model.add(Embedding(voc_size,dim,))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
